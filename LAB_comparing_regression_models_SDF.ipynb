Function for modeling

```python
def model_function(df, model_choice):
    data2_all = df.copy()
    y=data2_all['total_claim_amount']
    X=data2_all.drop(['total_claim_amount','customer'],axis=1)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=14)
    
    X_num_train = X_train.select_dtypes(include=np.number)
    X_num_test = X_test.select_dtypes(include=np.number)
    
    X_cat_train= X_train.select_dtypes(include=object)
    X_cat_test= X_test.select_dtypes(include=object)
    
    encoder = OneHotEncoder(handle_unknown='error', drop='first')
    encoder.fit(X_cat_train)
    X_cat_train_encoded = encoder.transform(X_cat_train).toarray()
    X_cat_test_encoded  = encoder.transform(X_cat_test).toarray()
    X_cat_train_encoded = pd.DataFrame(X_cat_train_encoded, columns = encoder.get_feature_names_out())
    X_cat_test_encoded  = pd.DataFrame(X_cat_test_encoded, columns = encoder.get_feature_names_out())
    
    X_train = pd.concat([X_num_train.reset_index(drop=True),X_cat_train_encoded],axis=1)
    X_test = pd.concat([X_num_test.reset_index(drop=True),X_cat_test_encoded],axis=1)
    
    
    StandardScalertransformer = StandardScaler()
    StandardScalertransformer.fit(X_train)
    X_train_stand = StandardScalertransformer.transform(X_train)
    X_test_stand = StandardScalertransformer.transform(X_test)
    
    X_train_stand = pd.DataFrame(X_train_stand, columns = X_train.columns)
    X_test_stand = pd.DataFrame(X_test_stand,  columns = X_test.columns)
    model = model_choice
    model.fit(X_train_stand,y_train)
    

    y_pred_train = model.predict(X_train_stand)
    y_pred_test = model.predict(X_test_stand)
    r2_score_train = r2_score(y_train,y_pred_train)
    r2_score_test = r2_score(y_test,y_pred_test)
    
    return r2_score_train, r2_score_test
```

```python
models = []
results_train = []
results_test = []
for k in range(2,10):
    models.append("Knn with {} neighbours".format(k))
    r1,r2 = model_function(data4,KNeighborsRegressor(n_neighbors=k,  weights='distance'))
    results_train.append(r1)
    results_test.append(r2)
    

models.append("Linear Model")
r1,r2 = model_function(data4,LinearRegression())
results_train.append(r1)
results_test.append(r2)

models.append("MLP Model")
r1,r2 = model_function(data4,MLPRegressor(random_state=14,max_iter=1000))
results_train.append(r1)
results_test.append(r2)

results_df = pd.DataFrame({'Model':models, 'R2_train':results_train, 'R2_test':results_test})
results_df
```
